```python
with open('Plot.out', 'r') as file:
    words = [ ]
    counts = [ ]
    for line in file:
        terms = line.rstrip().split(': ')
        words.append(terms[0])
        counts.append(int(terms[1]))
```

```python
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np
plt.rcdefaults()
labels = np.arange(len(counts))
plt.barh( labels, counts)
plt.yticks( labels + 0.5, words)
plt.xlabel( 'Count' )
plt.title( 'Top 20 (21 due to T-20th) Words in my Twitter Stream' )
plt.show()
```
